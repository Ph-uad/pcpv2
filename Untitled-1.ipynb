```python
#Data Collection
data = pd.read_csv('../data/data.csv')

#Data Cleaning
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)

#Data Validation
# print(data.info())
# print(data.describe())
assert data.notnull().all().all(), "Missing values present"
```

```python
def detect_mixed_columns(df):
    mixed_columns = []
    for col in df.columns:
        types = df[col].map(type).unique()
        if len(types) > 1:
            mixed_columns.append((col, types))
    return mixed_columns

# Identify mixed type columns
def coerce_mixed_columns():
    mixed_columns=[]

    for column in detect_mixed_columns(data):
            mixed_columns.append(column[0])
            
    for column in mixed_columns:
        data[column] = pd.to_numeric(data[column], errors='coerce')

data.dropna(inplace=True)
```
